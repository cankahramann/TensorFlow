### Feedforward Neural Networks for Regression

This notebook demonstrates the use of a feedforward neural network for regression analysis. This analysis will make use of synthetic data that we'll create for the analysis. The reason behind this is that this synthetic data is handy for understanding the behavior and the quirks of machine learning algorithms. This will be a good case to investigate where an algorithm fails or succeeds.

```python
# Import tensorflow
import tensorflow as tf

# Other imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Make the dataset
N = 1000
# The code below creates a uniformly distrubuted dataset bw. -3 and 3
X = np.random.random((N,2)) * 6 - 3
Y = np.cos(2*X[:,0] + np.cos(3*X[:,1]))
```

This will implement function:

> $y = \cos(2x_1) + \cos(3x_2)$

The reason behind implementing this functionality is that we want a graph with bumps and curves, and cosine helps us do that.

```python
# Plot it
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
plt.show()
```
